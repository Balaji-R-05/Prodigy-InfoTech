Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
population_data = pd.read_csv('population.csv')
print(population_data.shape)
print()
population_data.head(266)
```

```python
data_2023 = population_data.loc[:, ['Country Name', '2023']]
data_2023.dropna()
```

```python
print("First 5 rows")
data_2023.head()
```

```python
#List of contries for the graph.
countries_to_plot = ['United States', 'China', 'Brazil', 'India', 'Germany', 'France']

#Filtering the data frame
filtered_data = data_2023[data_2023['Country Name'].isin(countries_to_plot)]

#Creating the bar chart
plt.bar (filtered_data['Country Name'], filtered_data['2023'])
plt.xlabel ('Countries')
plt.ylabel ('Population (in Billions)')
plt.title('Population of selected countries in 2023')
plt.xticks(rotation=45, ha='right') #Rotating the x-axis labels for readability
plt.show()
```

```python

# Set up the visual style
sns.set(style="whitegrid")

# Histogram for population distribution in 2022
plt.figure(figsize = (12, 6))
sns.histplot(data = data_2023, x='2023', bins=30, kde=True, color='blue')
plt.title('Population Distribution in 2023')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()
```

```python
# Convert population column to numeric (in case it isn't already)
data_2023['2023'] = pd.to_numeric(data_2023['2023'], errors='coerce')

# Sort data by population in descending order and select top 10
top_10_countries = data_2023.sort_values(by='2023', ascending=False).head(10)

# Set up the visual style
sns.set(style="whitegrid")

# Bar chart for top 10 most populous countries in 2022
plt.figure(figsize=(12, 6))
sns.barplot(data = top_10_countries, x='Country Name', y='2023')
plt.title('Top Categorical Populations in 2023')
plt.xlabel('Regions')
plt.ylabel('Population (in billions)')
plt.xticks(rotation=50)
plt.show()
```

```python
# Example: Assume 'data' is the result of slicing
data = population_data.iloc[259, 4:68]

# Convert the Series to a DataFrame for easy manipulation (optional)
data = data.reset_index()
data.columns = ["Year", "Population"]

# Ensure Year is treated as a string or numeric for plotting
data["Year"] = data["Year"].astype(str)

# Plot the data
plt.figure(figsize=(12, 6))
plt.plot(data["Year"], data["Population"], marker="o", linestyle="-", color="blue")
plt.title("Population Over Time", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Population", fontsize=14)
plt.grid(True)
plt.xticks(rotation=90)  # Rotate x-axis labels for readability
plt.show()
```
